```python
# 1
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
def histogram_equalization(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    equalized = cv2.equalizeHist(img)

    cv2_imshow(img)
    cv2_imshow(equalized)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
# Example usage
histogram_equalization("/content/download.png")
```

```python
#2

import cv2
from google.colab.patches import cv2_imshow
import numpy as np
def thresholding(image_path, threshold_value=128):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    _, thresholded = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)

    cv2_imshow(img)
    cv2_imshow(thresholded)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage
thresholding('/content/download.png', threshold_value=128)
```

```python
#3

import cv2
import numpy as np
def edge_detection(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    edges = cv2.Canny(img, 100, 200)

    cv2_imshow(img)
    cv2_imshow(edges)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage
edge_detection('/content/download.png')
```

```python
#4

import cv2
import numpy as np

def data_augmentation(image_path):
    img = cv2.imread(image_path)

    # Example: Horizontal Flip
    flipped = cv2.flip(img, 1)

    cv2_imshow(img)
    cv2_imshow(flipped)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage
data_augmentation('/content/download.png')
```

```python
#5

import cv2
import numpy as np
def morphological_operations(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    kernel = np.ones((5, 5), np.uint8)
    erosion = cv2.erode(img, kernel, iterations=1)
    cv2_imshow(img)
    cv2_imshow(erosion)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
morphological_operations('/content/download.png')

```
