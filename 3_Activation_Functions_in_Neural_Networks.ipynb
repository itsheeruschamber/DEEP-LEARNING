```python
# 1 Sigmoid ACTIVATION FUNCTION
import numpy as np
def sig(x):
  return 1/(1+np.exp(-x))
x_val=np.linspace(-10,10,200)
y_val=sig(x_val)
import matplotlib.pyplot as plt
plt.plot(x_val,y_val,label='sigmoid')
plt.title('Sigmoid activation function')
plt.xlabel('Input')
plt.ylabel('output')
plt.legend()
plt.show()
```

```python
# 2 HYPERBOLIC TANGENT
import numpy as np
import matplotlib.pyplot as plt
def tanh(x):
    return np.tanh(x)
x_val = np.linspace(-10, 10, 200)
y_val = tanh(x_val)
plt.plot(x_val, y_val, label='tanh',color='Green')
plt.title('Hyperbolic Tangent (tanh) activation function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

```python
# 3 Rectified Linear Unit (ReLU) Activation Function:
import numpy as np
import matplotlib.pyplot as plt
def relu(x):
    return np.maximum(0, x)
x_val = np.linspace(-10, 10, 200)
y_val = relu(x_val)
plt.plot(x_val, y_val, label='ReLU')
plt.title('Rectified Linear Unit (ReLU) activation function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()

```

```python
#4 Leaky Rectified Linear Unit (Leaky ReLU) Activation Function:
import numpy as np
import matplotlib.pyplot as plt
def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)
x_val = np.linspace(-10, 10, 200)
y_val = leaky_relu(x_val)
plt.plot(x_val, y_val, label='Leaky ReLU (alpha=0.01)',color='Green')
plt.title('Leaky Rectified Linear Unit (Leaky ReLU) activation function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

```python
#5 Softmax Activation Function:
import numpy as np
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum(axis=0, keepdims=True)
logits = np.array([4.0, 2.0, 0.3])
probabilities = softmax(logits)
print("Logits:", logits)
print("Probabilities:", probabilities)
```
